### Scatter plot

Same as above but when explicitly defined, posses much more customizattion than generic plot method.

```python
# Generate two list of numbers / can also use numpy arrays
x = np.arange(10)
y = np.arange(10)
y = y*2 # Simply for fun

# View numbers
for i in range(10):
    print("Point {} is ({}, {})".format(i, x[i], y[i]))
```

```python
# Define a figure
plt.figure()

# Plot the points at once
# Set size of each point and color
plt.scatter(x, y, s=50, c="red")

# Show plot
plt.show()
```

```python
# Again lets refine the plot above

# Define a figure
plt.figure()

# Plot the points at once
# Set size of each point and color
plt.scatter(x, y, s=50, c="red")

# Add labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")

# Set axis ticks
plt.xticks(range(10))
plt.yticks(y)

# Show plot
plt.show()
```

```python
# Define a new figure
plt.figure()

# Plot multiple scatter plots at onces with diff sizes and colors
plt.scatter(np.array([2, 4, 6, 8]), np.array([2, 4, 6, 8]), s=100, color="skyblue")
plt.scatter(np.array([1, 5, 7, 1]), np.array([6, 4, 4, 3]), s=300, color="yellow")

# Add labels
plt.title("Multi-Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add legend... Yes LEGEND
plt.legend(["Plot-Name-1", "Plot-Name-2"], loc="upper right")

# Show plot
plt.show()
```

### Line plot

```python
# New figure
plt.figure()

# Some data to plot
y1 = np.arange(1, 10)

# Plot y using x as index 0, 1, ...
plt.plot(y1, "-o")

# Add labels
plt.title("Plot y-values using x as indexes")
plt.xlabel("x-index")
plt.ylabel("y-values")

# Set axis for x-axis and y-axis together
plt.axis([0, 10, 0, 10])

# Set ticks
plt.xticks(range(-5, 10))
plt.yticks(range(0, 10))

# Add sample legend
plt.legend(["Line1"])

# Show plot
plt.show()
```

```python
# New figure
plt.figure()

# Some data to plot
y1 = np.arange(1, 10)
y2 = np.array([1, 3, 5, 7, 9, 11, 13, 15, 17])

# Make sure number of points are same
assert len(y1) == len(y2)

# Plot y values using x as index
plt.plot(y1, "-o", y2, "-o")

# Add labels
plt.title("Area-under mulit-line plot")
plt.xlabel("Index 0, 1, 2, ...")
plt.ylabel("y-values")

# Set x and y ticks
plt.xticks(range(20))
plt.yticks(range(20))

# Fill the area between these two lines
plt.gca().fill_between(range(len(y1)), y1, y2, facecolor="blue", alpha=0.50)

# Add legend
plt.legend(["Plot Y1", "Plot Y2", "FullCover"])

# Show plot
plt.show()
```

### Bar Charts

### Creating subplots

```python
# New figure
plt.figure()


# Defne Subplot: multiple plots in one canvas

# Define 2 row, 2 col -> plot on index 1 out of 4
plt.subplot(2, 2, 1)

# Line plot
plt.plot(np.arange(10), "-o")

# Add labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Add title
plt.title("First-Row First-Column")

# Set ticks
plt.xticks(range(10))


# Define 2 row, 2 column -> plot on index 4
plt.subplot(2, 2, 4)

# Plot
plt.plot(np.arange(5), "-o")

# Add labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Add title
plt.title("First-Row Second-Column")

# Show
plt.show()
```

```python
# Share the x axis together and only change the y axis
fig, ((x1, x2), (x3, x4)) = plt.subplots(2, 2, sharex=True) # 4 plots: 2 rows and 2 columns

# All 4 x-axis
ax = [x1, x2, x3, x4]

for n in range(len(ax)):
    # Generate some data and plot
    size = 10**(n + 1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=size)
    
    # Plot histogram
    ax[n].hist(sample, bins=100)
    
    # Name the graphs
    ax[n].set_title("n = {}".format(size))
```

### Heatmaps

```python
# New figure canvas
plt.figure()

# Get somee random data: any normal or anything but make sure size is same
x = np.random.random(size=10000)
y = np.random.normal(loc=0.0, scale=1.0, size=10000)

# Plot a 2d histogram
# Basically heatmap - color changes based on the value
plt.hist2d(x, y, bins=100, cmap="hot")

# Add colorbar for reference
plt.colorbar()

# Show
plt.show()
```
